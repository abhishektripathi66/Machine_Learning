Normalization: it is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. it is also known as Min-Max scaling.

        Xnew = (Xi-min(X))/(max(x)-min(X))

here we would be using the min max technique
here the nature of the data doesnt changes after the normalization is performed,

the data is always present between the range of min and max, mostly the value is between 0 and 1

for example x -{2,3,4,5,6,7,8} min=2 max =7
based on this the new Xnew is calculated



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("../Data/loan_data.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min_max"]=ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_min_max"])
plt.plot()
```

we see that the nature is as is and the magniture has been altered thats all.
